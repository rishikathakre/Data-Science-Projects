## Data Preprocessing Pipeline

```python
# Step1 : Checking Missing Values
data.isnull().sum()
```

```python
# Step2 : Converting Categorical to Numerical
data['Species'].unique()
```

```python
data['Species']=data['Species'].map({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
```

```python
# Step3 : Check Outliers
plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1

for column in data:
    if plotnumber<=5 :
        ax = plt.subplot(3,2,plotnumber)
        sns.boxplot(x=data[column])
        plt.xlabel(column,fontsize=20)
        plt.ylabel('Species',fontsize=20)
    plotnumber+=1
plt.tight_layout()
```

```python
# Handling Outliers in SepalWidthCm
#Checking the distribution
sns.distplot(data.SepalWidthCm)
plt.show()
```

```python
# Interquartile range (IQR) 
from scipy import stats

IQR = stats.iqr(data.SepalWidthCm, interpolation = 'midpoint') 

IQR
```

```python
## min and max limit calculation
# First quartile (Q1) 
Q1 =data.SepalWidthCm.quantile(0.25)
# Third quartile (Q3) 
Q3 =data.SepalWidthCm.quantile(0.75)
##getting the limits
min_limit=Q1 - 1.5*IQR
print(min_limit)
max_limit=Q3 +1.5*IQR
print(max_limit)
```

```python
len(data.loc[data['SepalWidthCm'] < min_limit])
```

```python
len(data.loc[data['SepalWidthCm'] > max_limit])
```

```python
data.loc[data['SepalWidthCm'] > max_limit,'SepalWidthCm']=np.median(data.SepalWidthCm)
```

```python
data.loc[data['SepalWidthCm'] < min_limit,'SepalWidthCm']=np.median(data.SepalWidthCm)
```

```python
sns.boxplot(x=data['SepalWidthCm'])
plt.show()
```

```python
#Step4 : Checking Duplicate data
data.duplicated().sum()
```

```python
#Step5 : Scaling Data
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]=scale.fit_transform(data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])
```

## Feature Selection

```python
#Removing Unwanted/Constant Columns
l1=['Id']
data.drop(l1,axis=1,inplace=True)
```

```python
#Finding Correlation
corr_data=data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
sns.heatmap(corr_data.corr(),annot=True)
plt.show()
```

## Model Creation

```python
X= data.iloc[:,:-1]
y= data['Species']
```

```python
## Step 2 Creating training and testing data.
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=45)
```

```python
y_test
```

```python
## Step 3 Model creation
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(X_train,y_train)  ## training
```

```python
## Step 4 Prediction
y_pred= clf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,classification_report,f1_score
```

```python
# confusion matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
#recall score
recall=recall_score(y_test,y_pred,average='weighted')
recall
```

```python
#f1 score
f1score=f1_score(y_test,y_pred,average='weighted')
f1score
```

```python
#precision score
precision=precision_score(y_test,y_pred,average='weighted')
precision
```

```python
#classificatio report
cr=classification_report(y_test,y_pred)
print(cr)
```
